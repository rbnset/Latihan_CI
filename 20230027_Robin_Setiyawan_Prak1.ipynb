# Kondisional IF ... ELSE

image.png

```python
bilangan = int(input("Masukkan bilangan : "))

if(bilangan % 2 == 0) :
  print(bilangan, 'adalah bilangan GENAP')
else :
  print(bilangan, 'adalah  bilangan GANJIL')
```

# Kondisional IF ... ELIF ... ELSE

image.png

```python
bilangan = int(input("Masukkan bilangan : "))

if(bilangan > 0) :
  print(bilangan, "adalag bilangan POSITIF")
elif(bilangan < 0) :
  print(bilangan, "adalah bilangan NEGATIF")
else :
  print(bilangan, "adalah bilangan NETRAL")
```

# Perulangan FOR

image.png

```python
bilangan = int(input("Masukkan bilangan : "))

for i in range(bilangan) :
  print(i, end=' ')
```

```python
nama = input("Masukkan nama Anda : ")

for i in nama :
  print(i, end=' ')
```

# Perulangan WHILE

image.png

```python
bilangan = int(input("Masukkan bilangan : "))
i = 0

while(i < bilangan) :
  print(i, end=' ')
  i = i + 1
```

image.png

```python
nama = input("Masukkan nama Anda : ")
i = 0
while(i < len(nama)) :
  print(nama[i], end=' ')
  i = i + 1
```
